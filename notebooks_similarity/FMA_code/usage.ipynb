```python
import scipy.stats as stats

# Function to extract MFCC statistics from an MP3 file
def extract_mfcc_from_audio(y, sr):

    # Compute MFCCs (default is 13 coefficients)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)

    # Compute statistics
    mfcc_stats = {
        "mean": np.mean(mfcc, axis=1),
        "std": np.std(mfcc, axis=1),
        "min": np.min(mfcc, axis=1),
        "max": np.max(mfcc, axis=1),
        "median": np.median(mfcc, axis=1),
        "skew": stats.skew(mfcc, axis=1),
        "kurtosis": stats.kurtosis(mfcc, axis=1)
    }

    return mfcc_stats

# Function to extract MFCC values from a dataset (Pandas DataFrame)
def extract_mfcc_from_dataset(features, track_id):
    mfcc_df = features.loc[track_id, "mfcc"]
    return {stat: mfcc_df.loc[stat].values for stat in mfcc_df.index.levels[0]}

# Function to compare computed and dataset MFCCs
def compare_mfcc(features, track_id, y, sr):
    # Extract MFCCs from the dataset
    mfcc_dataset = extract_mfcc_from_dataset(features, track_id)

    # Compute MFCCs from the MP3 file
    mfcc_computed = extract_mfcc_from_audio(y, sr)

    # Compare and print differences
    for stat in mfcc_dataset.keys():
        dataset_values = mfcc_dataset[stat]
        computed_values = mfcc_computed.get(stat)

        if computed_values is not None:
            difference = np.abs(dataset_values - computed_values)
            print(f"{stat} - Mean Absolute Difference: {np.mean(difference):.5f}")
        else:
            print(f"{stat} not found in computed MFCCs")

# Example usage
track_id = 2  # The track ID to compare

# Assuming 'features' is your DataFrame
compare_mfcc(features, track_id, y, sr)
```

```python
# USE LIBROSA TO COMPUTE THE MFCC
# TODO - compare with obtained results
import scipy.stats as stats
def extract_mfcc_features(y, sr):

    # Compute MFCCs (default is 20 coefficients)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)

    # Compute statistics
    mfcc_mean = np.mean(mfcc, axis=1)  # Mean
    mfcc_std = np.std(mfcc, axis=1)  # Standard Deviation
    mfcc_min = np.min(mfcc, axis=1)  # Min
    mfcc_max = np.max(mfcc, axis=1)  # Max
    mfcc_median = np.median(mfcc, axis=1)  # Median
    mfcc_skew = stats.skew(mfcc, axis=1)  # Skewness
    mfcc_kurtosis = stats.kurtosis(mfcc, axis=1)  # Kurtosis

    # Store in dictionary
    mfcc_features = {
        "mean": mfcc_mean,
        "std": mfcc_std,
        "min": mfcc_min,
        "max": mfcc_max,
        "median": mfcc_median,
        "skew": mfcc_skew,
        "kurtosis": mfcc_kurtosis
    }

    return mfcc_features

# Example Usage
mfcc_stats = extract_mfcc_features(x, sr)

# Print result
for key, value in mfcc_stats.items():
    print(f"{key}: {value}")


# X = features.loc[small & (genre1 | genre2), 'mfcc']

```

## 5 Genre classification

### 5.1 From features

```python
small = tracks['set', 'subset'] <= 'small'

train = tracks['set', 'split'] == 'training'
val = tracks['set', 'split'] == 'validation'
test = tracks['set', 'split'] == 'test'

y_train = tracks.loc[small & train, ('track', 'genre_top')]
y_test = tracks.loc[small & test, ('track', 'genre_top')]
X_train = features.loc[small & train, 'mfcc']
X_test = features.loc[small & test, 'mfcc']

print('{} training examples, {} testing examples'.format(y_train.size, y_test.size))
print('{} features, {} classes'.format(X_train.shape[1], np.unique(y_train).size))
```

```python
# Be sure training samples are shuffled.
X_train, y_train = skl.utils.shuffle(X_train, y_train, random_state=42)

# Standardize features by removing the mean and scaling to unit variance.
scaler = skl.preprocessing.StandardScaler(copy=False)
scaler.fit_transform(X_train)
scaler.transform(X_test)

# Support vector classification.
clf = skl.svm.SVC()
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
print('Accuracy: {:.2%}'.format(score))
```
